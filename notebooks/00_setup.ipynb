```python
import pandas as pd

# ── Telecom: combine the 80/20 split back into one dataset ──
telecom_train = pd.read_csv(DATA + 'churn-bigml-80.csv')
telecom_test  = pd.read_csv(DATA + 'churn-bigml-20.csv')
telecom       = pd.concat([telecom_train, telecom_test], ignore_index=True)
print(f"Telecom shape: {telecom.shape}")

# ── ISP: load after unzip ─────────────────────────────────
# Replace filename below with what the unzip output showed you
isp = pd.read_csv(DATA + 'internet_service_churn.csv')
print(f"ISP shape: {isp.shape}")

# ── Insurance: combine train and test back into one ───────
# Replace filenames below with what the unzip output showed you
ins_train = pd.read_csv(DATA + 'Train.csv')
ins_test  = pd.read_csv(DATA + 'Test.csv')
insurance = pd.concat([ins_train, ins_test], ignore_index=True)
print(f"Insurance shape: {insurance.shape}")

datasets = {
    'Insurance': insurance,
    'ISP':       isp,
    'Telecom':   telecom
}

print("\n✓ All 3 datasets loaded successfully")
```
